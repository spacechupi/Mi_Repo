## Ejercicio 1
#### Crear un programa que calcule la cuota de una hipoteca

Las cuotas hipotecarias se calculan utilizando la fórmula del interés frances. El objetivo del ejercicio es crear un programa que solicite al usuario los datos de su hipoteca (cantidad, intereses, y plazo en años) y le devuelva la cuota mensual a pagar

La formula del interés frances es la siguiente:
\begin{align*}
C=V \frac{(1+i)^n i}{(1+i)^n -1}
\end{align*}


    C = Cuota a pagar
    V = Cantidad del préstamo hipotecario
    i = tipo de interés del período (tipo de interes de la hipoteca por mes)
    n = número de cuotas (en meses)

V = 300.000

i = 2 %

n = 20*12

output: 1517

````python
V = float(input("Cantidad de la hipoteca: "))
i = (float(input("Tipo de interés anual de la hipoteca: ")) / 12) / 100
n_anos = input("Plazo en años de la hipoteca: ")
n = int(n_anos) * 12

C = V * ((((1 + i) ** n) * i) / (((1 + i) ** n) - 1))

C = round(C, 2)

print (" ")
print("Su cuota mensual es de", C, "euros al mes.")
````

## Ejercicio 2
#### Crear un programa que calcule el plan de amortización

El objetivo del ejercicio es crear un programa que solicite al usuario los datos de su hipoteca (cantidad, intereses, y plazo en años) y le devuelva el plan de amortización.

En cada cuota se pagan intereses por el capital restante, y se devuelve capital hasta la cuota mensual. El capital restante ha de ser actualizado restandole el capital devuelto.

Salida ejemplo:

mes  cuota  intereses  principal  pendiente   

0  758.89   450.00   308.89   179691.11 

**NOTA**: Tienes que partir de tipo de interes por periodo calculado en el apartado anterior, y del total de la hipoteca. Para cada mes calcular:

1. cuota (apartado anterior)
2. intereses a pagar del mes (lo que quede por pagar * interes por periodo del apartado anterior)
3. principal (cuota - interes)
4. pendiente (lo que quede por pagar - principal)

````python
"""
Variables:

C = Cuota mensual
V = Cantidad del préstamo hipotecario
i = tipo de interés mensual
n = número de cuotas (en meses)
P = cantidad pagada

R = cantidad restante (pendiente)
i_mens = intereses a pagar del mes actual
p = principal
"""

V = float(input("Cantidad de la hipoteca: "))
i = (float(input("Tipo de interés anual de la hipoteca: ")) / 12) / 100
n_anos = input("Plazo en años de la hipoteca: ")
n = int(n_anos) * 12

# 1. Cuota:

C = V * ((((1 + i) ** n) * i) / (((1 + i) ** n) - 1))

C = round(C, 2)

print (" ")
print ("Plan de amortización de su hipoteca:")
print (" ")

P = 0
R = V

for mes in range(n):
    print("Mes:", mes)

# 1. Cuota mensual:
    print("Cuota mensual:", C)
    
# 2. Intereses del mes:
    i_mens = R * i
    i_mens = round(i_mens, 2)
    print("Interés a pagar del mes:", i_mens)
    
# 3. Principal:
    p = C - i_mens
    p = round(p, 2)
    print("Principal:", p)
    
# 4. Pendiente:
    R -= p
    print ("Pendiente:", R)
    print (" ")
    
    P += C
````

## Ejercicio 3
#### Crear un programa que resuelva una ecuacción de segundo grado

El objetivo del programa es resolver la ecuacción de sgundo grado 

\begin{align*}
ax^2+bx+c = 0
\end{align*}

Pasando como parámetros los coeficientes de la ecuacción. La resolución de la ecuacción es la siguiente:

\begin{align*}
x=\frac{-b \pm \sqrt{b^2-4ac}}{2a}
\end{align*}

Para realizar este programa tienes que utilizar el metodo sqrt() para calcular la raiz cuadrada. Para ello tienes que importar el módulo math. Así que tu programa debería empezar con ```ìmport math```


````python
import math

def ecuacion2o(a,b,c):
    disc = (b**2)-(4 * a * c)
    
    if disc > 0:
        x1 = ((-abs(b)) + math.sqrt(disc)) / (2 * a)
        x2 = ((-abs(b)) - math.sqrt(disc)) / (2 * a)
        return print("Dos soluciones reales y distintas:", x1, ",", x2)
    
    if disc == 0:
        x = ((-abs(b)) + math.sqrt(disc)) / (2 * a)
        return print("Una única solución real:", x)
    
    if disc < 0:
        return print("El discriminante es negativo. El resultado son dos números complejos.")
````

````python
print("x^2 + 5x + 6 = 0:")
ecuacion2o(1,5,6)
print (" ")

print("5x^2 + 2x - 3 = 0:")
ecuacion2o(5,2,-3)
print (" ")

print("x^2 + 2x + 3 = 0:")
ecuacion2o(1,2,3)
print (" ")

print("x^2 + 2x + 1 = 0:")
ecuacion2o(1,2,1)
print (" ")
````

````python
print("x^2 + 5x + 6 = 0:")
ecuacion2o_comp(1,5,6)
print (" ")

print("5x^2 + 2x - 3 = 0:")
ecuacion2o_comp(5,2,-3)
print (" ")

print("x^2 + 2x + 3 = 0:")
ecuacion2o_comp(1,2,3)
print (" ")

print("x^2 + 2x + 1 = 0:")
ecuacion2o_comp(1,2,1)
print (" ")
````

## Ejercicio 4
#### Crear un programa que calcule el average

El programa pedirá al usuario que introduzca numeros. En primer lugar le pedirá que indique cuantos números se quieren introducir, y enseñara un prompt para incluirlos, y calculará el average.

> ¿Cuántos números vas a introducir?: 3  
> Introduzca número: 10  
> Introduzca número: 20  
> Introduzca número: 30  
> La media es de 20  

````python
num_num = int(input("¿Cuántos números vas a introducir?: "))
num_num2 = num_num
sum_num = 0

while num_num2 > 0:
    num_input = int(input("Introduzca número: "))
    sum_num += num_input
    num_num2 -= 1

nums_avg = sum_num / num_num
nums_avg = int(nums_avg) # Esta línea sólo la he incluído porque el output de ejemplo no tiene punto decimal, /
                         # si quisiéramos que la media fuese un float se podría quitar o redondear con round().

print ("La media es de", nums_avg)
````

## Ejercicio 5
#### Crear un programa que calcule el average II

El programa pedirá que se diga cuantos números se quieren introducir, y enseñara un prompt para incluir número a número en una lista, y calcule después el average.


````python
num_num = int(input("¿Cuántos números vas a introducir?: "))
list_num = []

for num in range(num_num):
    num_input = int(input("Introduzca número: "))
    list_num.append(num_input)

nums_avg = sum(list_num) / num_num
nums_avg = int(nums_avg)

print ("La media es de", nums_avg)
````

## Ejercicio 7
#### Función de varios tipos
Define una función que reciba como argumentos un número entero y una lista, y devuelva los elementos de la lista multiplicados por el entero en formato lista.  

Ejecuta la función con los siguientes argumentos:
```Python
L = [1,2,3]
n = 5
```

Output esperado: 
```Python
[5, 10, 15]
```


````python
def mi_func(n, L):
    """
    Función que recibe como argumentos un número entero y una lista, /
    y devuelve los elementos de la lista multiplicados por el entero en formato lista. 
    
    Parámetros:
    n = número entero, tipo -> int
    L = lista de números -> list. Los elementos de la lista pueden ser de tipo int o float.
    """
    
    output = []
    
    for i in L:
        i = i * n
        output.append(i)
    
    return output
````

````python
mi_func(5, [1,2,3])
````
