```python
import random
```

```python
eu_mom = random.randint(0, 5)
print("eu mom:", eu_mom)

eu_dad = random.randint(0, 5)
print("eu dad:", eu_dad)

if eu_mom < eu_dad:
    eu = random.choice(range(eu_mom, eu_dad))
    print("eu:", eu)

if eu_mom == eu_dad:
    eu = eu_mom
    print("eu:", eu)
    
if eu_mom > eu_dad:
    eu = random.choice(range(eu_dad, eu_mom))
    print("eu:", eu)
```
